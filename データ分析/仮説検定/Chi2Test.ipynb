```python
import pandas as pd
import scipy as sp
import scipy.stats as stats
```

```python
df = pd.read_csv("golf.csv", encoding="shift-jis")
df
```

```python
cross = pd.crosstab(df.ゴルフ, [df.天気, df.風])
cross
```

```python
x2, p, dof, expected = sp.stats.chi2_contingency(cross)
```
