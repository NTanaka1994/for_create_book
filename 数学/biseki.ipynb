```python
#3Dグラフ
from mpl_toolkits.mplot3d import Axes3D
```

```python
coords=np.linspace(-4,4,500)
X,Y=np.meshgrid(coords,coords)
Z=np.cos(X*X+Y*Y)*0.95**(X*X+Y*Y)
fig,ax=plt.subplots(constrained_layout=True,subplot_kw={"projection":"3d"})
surf=ax.plot_surface(X,Y,Z,rstride=5,cstride=5,cmap="rainbow")
fig.colorbar(surf,ax=ax,shrink=0.6)
```

```python
plt.contourf(X,Y,Z,cmap="rainbow")
plt.show()
```

```python
import matplotlib.animation as anime
i=0
j=0
fig,ax=plt.subplots(constrained_layout=True,subplot_kw={"projection":"3d"})
def plot(data):
    coords=np.linspace(-4,4,200)
    X,Y=np.meshgrid(coords,coords)
    Z=np.cos((X+i)**2+(Y+j)**2)*0.95**(X*X+Y*Y)
    i=i+np.pi/10
    j=j+np.pi/10
    surf=ax.plot_surface(X,Y,Z,rstride=5,cstride=5,cmap="rainbow")
    fig.colorbar(surf,ax=ax,shrink=0.6)
anime.FuncAnimation(fig, plot, interval=100)
plt.show()
```
